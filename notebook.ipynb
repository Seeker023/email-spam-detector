# Email Spam Detection Walkthrough
This notebook covers Exploratory Data Analysis (EDA), Text Preprocessing, and Model Training for classifying emails as Spam or Ham.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('data/emails.csv')
df.head()
```

## Exploratory Data Analysis (EDA)

```python
print(f"Total emails: {len(df)}")
print(f"Missing values:\n{df.isnull().sum()}")

sns.countplot(x='spam', data=df)
plt.title('Distribution of Spam vs Ham')
plt.xticks([0, 1], ['Ham (0)', 'Spam (1)'])
plt.show()
```

```python
df['email_length'] = df['text'].astype(str).apply(len)

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='email_length', hue='spam', bins=50, kde=True)
plt.title('Email Length Distribution')
plt.xlim(0, 20000)
plt.show()
```
